# Задача 1

```python
def sum_distance(start, end): # from зарезервированные слово для импорта библиотек
  if start > end:
    start, end = end, start # Замена местами
  summa_chisel = 0
  for i in range(start, end + 1): # Пробегае от старта до энда
    summa_chisel = summa_chisel + i
  print(summa_chisel)
```

```python
sum_distance(1, 20)
```

```python
sum_distance(20, 1)
```

# Задача 2

```python
def trim_and_repeat(string: str, offset = 0, repetitions = 1):
  return(string[offset:] * repetitions)
```

```python
trim_and_repeat('Hi world', 3, 2)
```

```python
trim_and_repeat('Hi world', 6, 50)
```

```python
trim_and_repeat('Hellow world')
```
