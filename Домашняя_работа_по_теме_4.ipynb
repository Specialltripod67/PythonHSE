# Задание 1

```python
import json
purchases = {}
with open('purchase_log.txt', 'r', encoding='utf-8') as f: # encoding нужен для адекватного взаимодействия
  for i, line in enumerate(f):
    if i == 0: # скип заголовков
      continue
    labyda = json.loads(line) # парсим строку в словарь
    purchases[labyda["user_id"]] = labyda["category"] # заполняем словарь

for i, k in enumerate(purchases): # проходимся по ключам и выводи ключ = значение
  print(k, "'" + purchases[k] + "'")
  if i > 5:
    break
```

# Задание 2

```python
import csv
with open('visit_log.csv', 'r') as f:     # открываем файл
    with open('funnel.csv', 'w') as out:   # открываем файл на запись
      chitaem = csv.reader(f, delimiter = ';')  # создаём объект парсер
      writer = csv.writer(out)  # создаём объект для записи нашего файла
      fields = next(chitaem)  # скипаем заголовки
      headers = ["user_id", "source", "category"] # создаём заголовки для funnel
      writer.writerow(headers) # запись заголовков
      for i, values in enumerate(chitaem):     # цикл сравниваем
          user_id = values[0] # берём id из csv
          if user_id in purchases: # сравнение id
            category = purchases[user_id] # достаём категорию из словаря
            values.append(category)   # записываем что купили
            writer.writerow(values) # записываем всё в файл funnel.csv
```
