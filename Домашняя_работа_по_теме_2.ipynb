# Задача 1

```python
text_vvod = str(input('введите ваше слово: ')) #ввод слова
dlina = len(text_vvod) #узнаём его длину
if dlina % 2 == 0: #узнаём есть ли остаток от деления на 2
    rastoyanie_chet = dlina // 2 #узнаём индексы центральных букв
    print('word =', text_vvod, 'Результат:', text_vvod[rastoyanie_chet-1:rastoyanie_chet+1])
else:
    rastoyanie_ne_chet = dlina // 2 #узнаём индекс центральной буквы
    print('word =', text_vvod, 'Результат:', text_vvod[rastoyanie_ne_chet])
```

```python
def task2(boys, girls):
  if len(boys) != len(girls):
    print('Внимание, кто-то может остаться без пары!')
  else:
    boys=sorted(boys) #сортирует по алфавиту
    girls=sorted(girls) #сортирует по алфавиту
    i=0 #переменная для взятия нужного человека из списка
    print('идеальные пары:')
    for i in range(len(boys)): #соединяе все пары в диапозоне
      print(boys[i], 'и', girls[i])
```

```python
boys = ['peter', 'alex', 'john', 'arthur', 'richard', 'michael']
girls = ['kate', 'liza', 'kira', 'emma', 'trisha']
task2(boys, girls)
```
